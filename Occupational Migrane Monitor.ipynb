```python
def blood_pressure(df):
    Systolic = pd.Series([float(x[0]) for x in df['Blood Pressure'].str.split('/')], name='Systolic')
    Diastolic = pd.Series([float(x[1]) for x in df['Blood Pressure'].str.split('/')], name='Diastolic')
    return Systolic, Diastolic

def group_all(df):
    Gender_df = pd.DataFrame(df['Gender'])
    Age_df = pd.DataFrame(df['Age'])
    Occupation_df = pd.DataFrame(df['Occupation'])
    Sleep_Duration_df = pd.DataFrame(df['Sleep Duration'])
    Quality_of_Sleep_df = pd.DataFrame(df['Quality of Sleep'])
    Physical_Activity_Level_df = pd.DataFrame(df['Physical Activity Level'])
    Stress_Level_df = pd.DataFrame(df['Stress Level'])
    BMI_Category_df = pd.DataFrame(df['BMI Category'])
    Systolic_df, Diastolic_df = blood_pressure(df)
    Heart_Rate_df = pd.DataFrame(df['Heart Rate'])
    Daily_Steps_df = pd.DataFrame(df['Daily Steps'])
    Sleep_Disorder_df = pd.DataFrame(df['Sleep Disorder'])

    cleaned_df = pd.concat([Gender_df, Age_df, Occupation_df, Sleep_Duration_df, Quality_of_Sleep_df,Physical_Activity_Level_df,
                            Stress_Level_df, BMI_Category_df, Systolic_df,Diastolic_df, Heart_Rate_df, Daily_Steps_df,
                            Sleep_Disorder_df], axis=1)
    return cleaned_df.copy()
```

```python
final_df = group_all(df)
final_df
```

## Predicting Headache Risk

```python
def label_headache(row):
  """Labels headache based on stress, sleep quality, and duration."""
  if row['Stress Level'] > 7 and row['Quality of Sleep'] < 6 and row['Sleep Duration'] < 6:
    return "High Risk of Headache"
  elif row['Stress Level'] > 7 or row['Quality of Sleep'] < 6 or row['Sleep Duration'] < 6:
      return "Moderate Risk of Headache"
  else:
    return "Low Risk of Headache"

# Apply the function to create the 'Headache Label' column
final_df['Headache Label'] = final_df.apply(label_headache, axis=1)

# Display the DataFrame with the new column
print(final_df[['Stress Level', 'Quality of Sleep', 'Sleep Duration', 'Headache Label']])
```
